```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Find unique players by applying valuecounts to the column "SN" and count the total number of players
Total_Players = len(purchase_data["SN"].unique())

#Create a data frame displaying the total number of players
Total_Players = pd.DataFrame({"Total Players":[Total_Players]})
Total_Players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Find a series containing all unique items and the total number
unique_item = len(purchase_data["Item ID"].unique())

#Find the average price and round to 2 decimals
average_price = round(purchase_data["Price"].mean(),2)

#Find the total number of purchases made
number_of_purchase = purchase_data["Purchase ID"].count()

#Find the revenue by summing all the prices in the dataframe
total_revenue = purchase_data["Price"].sum()

#Construct a data frame containing all the information above
summary = pd.DataFrame({
    "Number of unique Items":[unique_item],
    "Average Price":["$"+str(average_price)],
    "Number of Purchases":[number_of_purchase],
    "Total Revenue":["$"+str(total_revenue)]
})
summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Group the dataframe by gender
purchase_gender = purchase_data.groupby("Gender")

#Find the distinct "SN"s in the dataframe
gender_co = purchase_gender.nunique()["SN"]

#Find the sum of players of all genders
sum_counts = gender_co.sum()

#Construct a dataframe of gender demographics
gender_counts = pd.DataFrame({"Total Count":gender_co,
                            "Percentage of Players":gender_co/sum_counts})
#Change the format of the percentage column
gender_counts["Percentage of Players"]=(gender_counts["Percentage of Players"]*100).map("{:.2f}%".format)

#Sort the data frame from highest count to lowest
gender_counts = gender_counts.sort_values("Total Count",ascending = False)
gender_counts
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Set bins and age group names for the data given
bins = [0,10,15,20,25,30,35,40,9999]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Categorize using age bins
age_bins = pd.cut(purchase_data["Age"], bins, right = False, labels = group_names)

#copy the original dataframe and add in the column cotaining age bins
purchase_age = purchase_data.copy()
purchase_age["age summary"]=age_bins

#Calculate the numbers and percentages by age group
purchase_age = purchase_age.groupby("age summary")
age_count = purchase_age.nunique()["SN"]
age_count_percent = age_count/age_count.sum()

#Construct a data frame of age demographics
age_demographics = pd.DataFrame({"Total Count":age_count,
                                "Percentage of Players":age_count_percent})
age_demographics.index.name = None

#Change the format of percentage
age_demographics["Percentage of Players"]=(age_demographics["Percentage of Players"]*100).map("{:.2f}%".format)
age_demographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group the dataframe by "SN"
purchase_person = purchase_data.groupby("SN")

#Calculate the total purchase value for each person
total_price = purchase_person["Price"].sum()

#Calculate the total purchase for each person in amount
total_count = purchase_person["Purchase ID"].count()

#Calculate the average purchase price for each person
avg_price = total_price/total_count

#Construct a data frame for top spenders
top_spenders = pd.DataFrame({"Purchase Count":total_count,
                             "Average Purchase Price":avg_price,
                             "Total Purchase Value":total_price})

#Change the format
top_spenders = top_spenders.sort_values("Total Purchase Value", ascending = False)
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:.2f}".format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:.2f}".format)

#Preview of the summary data frame
top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve the 3 columns
most_popular = purchase_data.loc[:,["Item Name","Item ID","Price"]]

#Group by item ID and item name
most_popular = most_popular.groupby(["Item ID","Item Name"])

#Calculate the purchase count
purchase_count_item = most_popular["Price"].count()

#Calculate the item price
price_item = most_popular["Price"].unique()

#Calculate the total purchase value
total_purchase_value = most_popular["Price"].sum()

#Construct a data frame of most popuat items
most_popular_item = pd.DataFrame({"Purchase Count":purchase_count_item,
                                  "Item Price":price_item,
                                  "Total Purchase Value":total_purchase_value})

#Formatting
most_popular_items = most_popular_item.sort_values("Purchase Count", ascending = False)
#Change the column from string to float
most_popular_items["Item Price"] =most_popular_items["Item Price"].astype(float)
most_popular_items["Item Price"] = most_popular_items["Item Price"].map("${:,}".format)
most_popular_items["Total Purchase Value"] = most_popular_items["Total Purchase Value"].map("${:.2f}".format)

most_popular_items.head()
```
